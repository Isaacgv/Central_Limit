# Central Limit Theorem
Simulating several toss of a coin

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
from  matplotlib.animation import FuncAnimation
%matplotlib notebook
sns.set()
```

```python
#Return the list of the tanking a random sample of the several toss of a coin
def central_limit(N, size_sample, number_sample):
    n = 1 #Number of trials
    p = .5 #Probability of each trial

    #Simulating several toss of a coin
    L1 = np.random.binomial(n, p, N)

    #Taking a sample of random results among L1 
    L2 = [np.random.choice(L1, size_sample, replace=False).mean() for n in range(number_sample)]
    
    return L2
```

```python
#Plot the number of samples 1 to number_sample (Number of samples)
def update_hist(number_sample, size_sample, N):
    
    plt.cla()
    
    #Obtain the samples of the toss of a coin
    L2 = central_limit(N, size_sample=size_sample, number_sample=number_sample)

    #Plot histogram
    sns.histplot(L2, kde=True)
    
    plt.ylabel("Count")
    plt.xlabel("Mean")
    
    mean = np.mean(L2)
    median = np.median(L2)
    mode = stats.mode(L2)[0]
    
    for element_mode in mode:
        plt.axvline(element_mode, color='b', linestyle='--', linewidth=1, label ="Mode")
        
    plt.title("Number of Size={} Mean={:.2f} Median={:.2f} Moda={}".format(number_sample+1, mean, median, mode)) 
    plt.axvline(mean, color='k', linestyle='--', linewidth=1, label ="Mean")
    plt.axvline(median, color='r', linestyle='--', linewidth=1, label ="Median")
    
    plt.legend()
    
```

```python
N = int(input("Number of toss of a coin: "))
size_sample = int(input("Size of the sample: "))
number_sample = int(input("Number of samples: "))

number_of_frames = number_sample
fig = plt.figure()
animation = FuncAnimation(fig, update_hist, number_of_frames, interval=200, repeat=False, 
                          fargs=(number_sample, N))
plt.show()
```

## Annotations 
- The sampling distribution of the mean will be less spread than the values in the population from which the sample is draw.
- The sampling distribution will modeled by a normal distribution as the sample size increases.
- The spread of the sampling distribution is related to the spread of the population values.
- Bigger samples lead to  a smaller spread in the sampling distribution.
- The distribution of the sample mean tends to the normal distribution as the sample size increases regardless of the sampling distribution.
